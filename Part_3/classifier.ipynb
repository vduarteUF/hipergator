```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.mdlstm1 = tm.layers.BDHW_LSTM(2)
        # First 2D convolutional layer, taking in 1 input channel (image),
        # outputting 32 convolutional features, with a square kernel size of 3
        self.conv1 = nn.Conv2d(1, 6, 2, 4)
        self.tanh1 = nn.tanh()
        
        # Second 2D convolutional layer, taking in the 32 input layers,
        # outputting 64 convolutional features, with a square kernel size of 3
        self.mdlstm2 = tm.layers.BDHW_LSTM(10)
        self.conv2 = nn.Conv2d(1, 20, 2, 4)
        self.tanh2 = nn.tanh()

        self.mdlstm3 = tm.layers.BDHW_LSTM(50)

        # First fully connected layer
        self.fc1 = nn.Linear(50, 80)
        self.softmax1 = nn.softmax()
        
    def forward(self, x):
        x = self.mdlstm1(x)
        x = self.conv1(x)
        x = self.sum(x, dim=1)
        x = self.tanh1(x)
        x = self.mdlstm2(x)
        x = self.conv2(x)
        x = self.sum(x, dim=1)
        x = self.tanh2(x)
        x = self.mdlstm3(x)
        x = self.fc1(x)
        x = self.softmax1(x)
        
```
